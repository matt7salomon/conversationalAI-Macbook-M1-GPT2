```python
import torch
from transformers import GPT2Tokenizer, GPT2LMHeadModel
import speech_recognition as sr
import pyttsx3
import sounddevice as sd
import numpy as np

# Initialize the text-to-speech engine
engine = pyttsx3.init()

# Set speech rate (optional)
engine.setProperty('rate', 150)

# Check if we can use the GPU on a Mac M1
device = torch.device("mps") if torch.backends.mps.is_available() else torch.device("cpu")
print(f"Using device: {device}")

# Load the GPT-2 tokenizer and model
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2").to(device)
model.eval()

# Speech recognition setup
recognizer = sr.Recognizer()

# Function to convert text to speech
def speak(text):
    engine.say(text)
    engine.runAndWait()

# Function to listen for user input via microphone
def listen():
    with sr.Microphone() as source:
        recognizer.adjust_for_ambient_noise(source)
        print("Listening...")
        audio = recognizer.listen(source)

        try:
            # Recognize the speech
            user_input = recognizer.recognize_google(audio)
            print(f"User: {user_input}")
            return user_input
        except sr.UnknownValueError:
            print("Sorry, I could not understand your speech.")
            return ""
        except sr.RequestError as e:
            print(f"Could not request results from Google Speech Recognition service; {e}")
            return ""

# Function to generate a response using GPT-2
def generate_response(prompt, max_length=100):
    inputs = tokenizer.encode(prompt, return_tensors="pt").to(device)
    outputs = model.generate(inputs, max_length=max_length, pad_token_id=tokenizer.eos_token_id)
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response

# Main conversation loop
def conversation():
    print("Say something to start the conversation (or 'quit' to exit):")
    while True:
        user_input = listen()
        
        if user_input.lower() == "quit":
            print("Exiting...")
            speak("Goodbye!")
            break

        if user_input:
            # Generate AI response
            gpt_response = generate_response(user_input)
            print(f"AI: {gpt_response}")
            
            # Convert response to speech
            speak(gpt_response)

if __name__ == "__main__":
    conversation()
```

## Voice response from the LLM this time with the GTTS module

```python
import torch
from transformers import GPT2Tokenizer, GPT2LMHeadModel
import speech_recognition as sr
from gtts import gTTS
import playsound
import os

# Check if we can use the GPU on a Mac M1
device = torch.device("mps") if torch.backends.mps.is_available() else torch.device("cpu")
print(f"Using device: {device}")

# Load the GPT-2 tokenizer and model
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2").to(device)
model.eval()

# Speech recognition setup
recognizer = sr.Recognizer()

# Function to convert text to speech using gTTS and play it
def speak(text):
    tts = gTTS(text=text, lang='en')
    filename = "temp_audio.mp3"
    tts.save(filename)
    playsound.playsound(filename)
    os.remove(filename)

# Function to listen for user input via microphone
def listen():
    with sr.Microphone() as source:
        recognizer.adjust_for_ambient_noise(source)
        print("Listening...")
        audio = recognizer.listen(source)

        try:
            # Recognize the speech
            user_input = recognizer.recognize_google(audio)
            print(f"User: {user_input}")
            return user_input
        except sr.UnknownValueError:
            print("Sorry, I could not understand your speech.")
            return ""
        except sr.RequestError as e:
            print(f"Could not request results from Google Speech Recognition service; {e}")
            return ""

# Function to generate a response using GPT-2
def generate_response(prompt, max_length=100):
    inputs = tokenizer.encode(prompt, return_tensors="pt").to(device)
    outputs = model.generate(inputs, max_length=max_length, pad_token_id=tokenizer.eos_token_id)
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response

# Main conversation loop
def conversation():
    print("Say something to start the conversation (or 'quit' to exit):")
    while True:
        user_input = listen()
        
        if user_input.lower() == "quit":
            print("Exiting...")
            speak("Goodbye!")
            break

        if user_input:
            # Generate AI response
            gpt_response = generate_response(user_input)
            print(f"AI: {gpt_response}")
            
            # Convert response to speech
            speak(gpt_response)

if __name__ == "__main__":
    conversation()
```
